```sql
CREATE OR REPLACE TEMP VIEW dashboard_kpi AS
SELECT
    COUNT(DISTINCT game_id) AS total_games,
    COUNT(*) AS total_records,
    ROUND(AVG(net_worth),2) AS average_net_worth,
    ROUND(AVG(total_properties),2) AS average_properties
FROM workspace.default.monopoly_clean;
```

```sql
CREATE OR REPLACE TEMP VIEW dashboard_visits AS
SELECT
    position,
    COUNT(*) AS visits
FROM workspace.default.monopoly_clean
GROUP BY position
ORDER BY position;
```

```sql
CREATE OR REPLACE TEMP VIEW dashboard_properties AS
SELECT
    total_properties,
    ROUND(AVG(net_worth),2) AS average_net_worth
FROM workspace.default.monopoly_clean
GROUP BY total_properties
ORDER BY total_properties;
```

```sql
CREATE OR REPLACE TEMP VIEW dashboard_turns AS
SELECT
    turn,
    ROUND(AVG(net_worth),2) AS average_net_worth
FROM workspace.default.monopoly_clean
GROUP BY turn
ORDER BY turn;
```

```sql
CREATE OR REPLACE TEMP VIEW dashboard_monopolies AS
SELECT
    monopolies,
    ROUND(AVG(net_worth),2) AS average_net_worth
FROM workspace.default.monopoly_clean
GROUP BY monopolies
ORDER BY monopolies;
```
