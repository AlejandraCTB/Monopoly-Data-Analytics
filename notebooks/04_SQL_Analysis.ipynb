```sql
SELECT
    COUNT(*) AS total_records,
    COUNT(DISTINCT game_id) AS total_games,
    COUNT(DISTINCT player_number) AS total_players,
    MAX(turn) AS max_turns
FROM workspace.default.monopoly_clean;
```

```sql
SELECT
    position,
    COUNT(*) AS visits
FROM workspace.default.monopoly_clean
GROUP BY position
ORDER BY visits DESC;
```

```sql
SELECT
    position,
    ROUND(AVG(net_worth),2) AS avg_net_worth
FROM workspace.default.monopoly_clean
GROUP BY position
ORDER BY avg_net_worth DESC;
```

```sql
SELECT
    monopolies,
    COUNT(*) AS observations,
    ROUND(AVG(net_worth),2) AS average_net_worth
FROM workspace.default.monopoly_clean
GROUP BY monopolies
ORDER BY monopolies;
```

```sql
SELECT
    total_properties,
    ROUND(AVG(net_worth),2) AS avg_net_worth,
    COUNT(*) AS observations
FROM workspace.default.monopoly_clean
GROUP BY total_properties
ORDER BY total_properties;
```

```sql
SELECT
    houses,
    hotels,
    ROUND(AVG(net_worth),2) AS avg_net_worth
FROM workspace.default.monopoly_clean
GROUP BY houses, hotels
ORDER BY houses, hotels;
```

```sql
SELECT
    jail_count,
    ROUND(AVG(net_worth),2) AS avg_net_worth,
    COUNT(*) AS observations
FROM workspace.default.monopoly_clean
GROUP BY jail_count
ORDER BY jail_count;
```

```sql
SELECT
    has_lost,
    COUNT(*) AS observations,
    ROUND(AVG(net_worth),2) AS avg_net_worth,
    ROUND(AVG(total_properties),2) AS avg_properties,
    ROUND(AVG(monopolies),2) AS avg_monopolies
FROM workspace.default.monopoly_clean
GROUP BY has_lost;
```

```sql
SELECT
    turn,
    ROUND(AVG(net_worth),2) AS avg_net_worth
FROM workspace.default.monopoly_clean
GROUP BY turn
ORDER BY turn;
```

```sql
SELECT
    game_id,
    player_number,
    turn,
    net_worth,
    total_properties,
    monopolies,
    houses,
    hotels
FROM workspace.default.monopoly_clean
ORDER BY net_worth DESC
LIMIT 20;
```
